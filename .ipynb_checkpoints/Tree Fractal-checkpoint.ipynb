```python
import numpy as np
import pylab as pl
from matplotlib import collections  as mc
import math
```

```python
# for (x1,y1) and (x2,y2) find euclidean distance
def distance(a,b):
    return math.sqrt((b[0]-a[0])**2+(b[1]-a[1])**2)
```

```python
angle = math.pi/2
deflect = math.pi/4
shrink = 2/3
branch=[]
trunk= [[(0,0),(0,1)]]
endpoints = {trunk[0][1]:angle}
trunklength = distance(trunk[0][0],trunk[0][1])
for i in range(9):
    startpoints=endpoints
    endpoints = {}
    for startpoint in startpoints.keys(): 
            x1 = startpoint[0]+(shrink)**i*trunklength*math.cos(startpoints[startpoint]+deflect)
            y1 = startpoint[1]+(shrink)**i*trunklength*math.sin(startpoints[startpoint]+deflect)
            endpoints[(x1,y1)]=startpoints[startpoint]+deflect
            
            x2= startpoint[0]+(shrink)**i*trunklength*math.cos(startpoints[startpoint]-deflect)
            y2=startpoint[1]+(shrink)**i*trunklength*math.sin(startpoints[startpoint]-deflect)
            endpoints[(x2,y2)]=startpoints[startpoint]-deflect
            
            
            branch.append([startpoint,(x1,y1)])
            branch.append([startpoint,(x2,y2)])
```

```python
tree = trunk + branch
lc = mc.LineCollection(tree, linewidths=2)
fig, ax = pl.subplots()
ax.add_collection(lc)
ax.autoscale()
ax.margins(0.1)
ax.set_aspect('equal', adjustable='box')
```
